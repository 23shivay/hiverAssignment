```python


UNIVERSAL_CATEGORIES = [
    "access_issue",
    "workflow_issue",
    "threading_issue",
    "tagging_issue",
    "ui_bug",
    "automation_issue",
    "auth_issue",
    "export_issue",
    "notification_issue",
    "feature_request",
    "billing_issue",
    "analytics_issue",
    "performance_issue",
    "mobile_issue",
    "sla_issue",
    "assignment_issue",
    "sync_issue",
    "editor_issue",
    "attachment_issue",
    "user_management_issue",
    "forwarding_issue",
    "signature_issue",
    "ui_state_issue",
    "suggestion_issue",
    "information_request",
    "duplication_issue",
    "logging_issue",
    "session_issue",
    "search_issue",
    "send_issue",
    "setup_help",
    "other_issue"
]


def classify_category(customer, subject, body):
    prompt = f"""
You are an email classifier for customer support tickets.

Classify the email into EXACTLY one of these categories:
{UNIVERSAL_CATEGORIES}

Rules:
- Use both subject and body.
- Pick the most accurate universal category.
- Only choose "other_issue" if none of the categories fit.

Email:
Customer: {customer}
Subject: {subject}
Body: {body}

Return ONLY the category name.
"""

    resp = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=15,
        temperature=0
    )

    return resp.choices[0].message.content.strip()
```
